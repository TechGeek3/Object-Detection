```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_file="ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
frozen_model="frozen_inference_graph.pb"
```

```python
model=cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
classLabel=[]   #empty list
file_name="Labels.txt"
with open(file_name,'rt') as f:
    classLabel=f.read().rstrip('\n').split('\n')  #appending classLabel
```

```python
print(classLabel)   #printing all the objects that can be detected using this model
```

```python
print(len(classLabel))   #total no. of possible detected objects
```

```python
model.setInputSize(320,320)   #from config file
model.setInputScale(1.0/127.5)   #255/2
model.setInputMean((127.5,127.5,127.5))   #mobilenet uses [-1,1]
model.setInputSwapRB(True)   #for automatic swapping of colors
```

## Detecting using video

```python
cap=cv2.VideoCapture("Traffic.mp4")   #Reading video frames
cap.set(3,650)
cap.set(4,450)

while True:
    success,web=cap.read()
    
    classIndex,confidence,bbox=model.detect(web,confThreshold=0.5)   #bbox-bounding box
    print(classIndex)
    
    if(len(classIndex)!=0):
        for classInd,conf,boxes in zip(classIndex.flatten(),confidence.flatten(),bbox):   #zip-for 3 parameters,flatten()-to convert to 1d array
            if(len(classIndex)<=80):
                cv2.rectangle(web,boxes,(0,255,0),3)
                cv2.putText(web,classLabel[classInd-1].upper(),(boxes[0]+10,boxes[1]+40),cv2.FONT_HERSHEY_PLAIN,3,(255,0,0),2)
                
                #cv2.putText(web,str(round(conf*100,2)),(boxes[0]+200,boxes[1]+40),cv2.FONT_HERSHEY_PLAIN,3,(255,0,0),2)
     
    cv2.imshow("Object Detection",web)
    
    if cv2.waitKey(1) & 0xff==ord('q'):
        break
        
cap.release()     
cv2.destroyAllWindows()
```
